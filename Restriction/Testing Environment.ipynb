```python
import tkinter as tk
from tkinter import *

calibrate = False # While the exact z_coord isn't set, keep running the loop program
while calibrate == False:
    window = tk.Tk() #open window
    
    def confirm(): # If confirm button pressed, destory window and change calibrated to true
        window.destroy()
        global calibrate
        calibrate = True
        
        
    def test(): #If test button pressed, grab z variable from entry and run in to robot
        global z_coord
        z_coord = z_entry.get()
        window.destroy()
        
        
    warning = tk.Label(text="Start with a large value and slowly increment down until optimal z coord")
    warning.pack()
    
    z_entry = tk.Entry()
    z_entry.pack()
    
    
    test_button = tk.Button(text="Test",command=test)
    test_button.place(rely=0.2,relx=0.4)
    
    confirm_button = tk.Button(text="Confirm",command=confirm)
    confirm_button.place(rely=0.2,relx=0.6)

    window.geometry("400x400")
    window.title("Calibration Software")
    window.mainloop()

z_coord
```

```python
import tkinter as tk
from tkinter import *



def set_enzyme(entry):   # Takes entry and saves into variable
    enzyme_name = entry
    return enzyme_name

def exit():  # exits the window
    root.destroy()
    return

root = tk.Tk()  # creates window
root.geometry("400x400") # sets geometry
root.title("Parameter Selection") # titles the window

enzyme_name = tk.StringVar()

enzyme_entry = tk.Entry(root,textvariable = enzyme_name) #creates an entry box -> places it in "root" window
enzyme_entry.pack()

enzyme_label = tk.Label(text="Type in your enzyme exactly") #creates a label -> places it in "root" window
enzyme_label.pack()

exit_button = tk.Button(text = "Exit",command=exit)
exit_button.pack()

confirm_button = tk.Button(text = "Confirm Enzyme",command=set_enzyme(str(enzyme_entry.get())))
confirm_button.pack()

             
root.mainloop()  #looping the way the window responds to changes in environment
```

```python
root=tk.Tk()
 
# setting the windows size
root.geometry("600x400")
  
# declaring string variable
# for storing name and password
name_var=tk.StringVar()
passw_var=tk.StringVar()
 
  
# defining a function that will
# get the name and password and
# print them on the screen
def submit():
 
    name=name_var.get()
    password=passw_var.get()
     
    print("The name is : " + name)
    print("The password is : " + password)
     
    name_var.set("")
    passw_var.set("")
     
     
# creating a label for
# name using widget Label
name_label = tk.Label(root, text = 'Username', font=('calibre',10, 'bold'))
  
# creating a entry for input
# name using widget Entry
name_entry = tk.Entry(root,textvariable = name_var, font=('calibre',10,'normal'))
  

passw_label = tk.Label(root, text = 'Password', font = ('calibre',10,'bold'))
  

passw_entry=tk.Entry(root, textvariable = passw_var, font = ('calibre',10,'normal'), show = '*')
  

sub_btn=tk.Button(root,text = 'Submit', command = submit)
  

name_label.grid(row=0,column=0)
name_entry.grid(row=0,column=1)
passw_label.grid(row=1,column=0)
passw_entry.grid(row=1,column=1)
sub_btn.grid(row=2,column=1)
  

root.mainloop()
```
